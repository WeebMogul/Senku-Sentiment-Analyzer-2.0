# Data Cleaning Process

## Step 1: View data

```python
import pandas as pd

df = pd.read_csv('D:\Github Projects\Heriot-Watt-Msc-Project-Sentiment-Analysis\Comment_Extractor\Kaguya-sama\Kaguya_sama_Episode_2_Comment_list.csv')
df.head()
```

## Step 2: Remove the deleted and removed tags

- ### 2.1. Deleted comment removal

```python
deletetag_index = df[df['Comment'] == '[deleted]'].index
deletetag_index
```

```python
df.drop(deletetag_index,inplace=True)
df
```

- ### 2.2 Removed comment tags removal

```python
removetag_index = df[df['Comment'] == '[removed]'].index
df.drop(removetag_index,inplace=True)
df
```

## Step 3: Remove tags for spoiler and links

- ### 3.1 : Hyperlink tags

Remove the website links from the comments using the regex : '\(https?:\/\/.*[\r\n\)]*\)'

```python
df['Comment']= df['Comment'].str.replace(r'\(https?:\/\/.*[\r\n\)]*\)','')
df
```

- ### 3.2 : Memearrows,newline and \r characters

Remove the Memearrows,newline and \r characters using the following regex ; '\>|\/r|\/n'

```python
df['Comment']= df['Comment'].str.replace(r'\>',' ')
df['Comment']= df['Comment'].str.replace('\r',' ')
df['Comment']= df['Comment'].str.replace('\n',' ')
df
```

- ### 3.3 : Brackets 

Remove the brackets using the following regex expression : '\[|\]|\)'

```python
df['Comment']= df['Comment'].str.replace(r'\[|\]|\)','')
df
```

- ### 3.4 : Unnecessary characters and spoiler tags

Remove spoiler tags using '\(\/s.' and unnecessary characters using '\*|\~|\#')

```python
df['Comment']= df['Comment'].str.replace(r'\(\/s.',' : ')
df['Comment']= df['Comment'].str.replace(r'\*|\~|\#|\\',' ')
df['Comment']= df['Comment'].str.replace('&x200B',' ')
df
```
